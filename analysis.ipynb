```python
import requests
import json
import pandas as pd
import seaborn as sb
import numpy as np
```

```python
#Crime Statistics CSV
crimeDF = pd.read_csv("report.csv")
```

```python
crimeDF
```

# Denver, CO 
# Yearly Crime Per Capita

```python
#denverCrimeDF is for calculating crimes_percapita of just denver
#i call crimeDF and use sort_values and query to find denver co and sort by year
denverCrimeDF = crimeDF[["report_year","agency_jurisdiction","crimes_percapita"]].sort_values("report_year",
            ascending=True).query("agency_jurisdiction=='Denver, CO'")
denverCrimeDF
```

```python
#i used a seaborn catplot to show the crime over the years x as years and y as crime pc
sb.catplot(data=denverCrimeDF,x='report_year',y='crimes_percapita',aspect=2,kind="point",hue="agency_jurisdiction",palette='Blues').set_xticklabels(rotation=90)
```

# Jacksoville, FL 
# Yearly Crime Per Capita


```python
#i did the same as above but queried by Jacksonville Fl
jaxCrimeDF = crimeDF[["report_year","agency_jurisdiction","crimes_percapita"]].sort_values("report_year",
            ascending=True).query("agency_jurisdiction=='Jacksonville, FL'")
jaxCrimeDF
```

```python
#created a catplot, x report year and y crimes pc, palette set to oranges
sb.catplot(data=jaxCrimeDF,x='report_year',y='crimes_percapita',aspect=2,kind="point",hue="agency_jurisdiction",palette='Oranges').set_xticklabels(rotation=90)
```

# Denver, CO and Jacksoville, FL 
# Yearly Total Violent Crime 

```python
#to aggregate the data I queried Jax and Den 
jaxDenverCrimeDF = crimeDF.query("agency_jurisdiction=='Denver, CO' or agency_jurisdiction=='Jacksonville, FL'")

#I grouped them by year and city. I aggregated by the sum of crimes_percapita
groupedJaxDenverCrimeDF = jaxDenverCrimeDF.groupby(["report_year","agency_jurisdiction"])
aggGroupedJaxDenverCrimeDF = groupedJaxDenverCrimeDF.agg({"crimes_percapita":sum})

#i reset the index to clean up the data
resetAggGroupedJaxDenverCrimeDF = aggGroupedJaxDenverCrimeDF.reset_index()

resetAggGroupedJaxDenverCrimeDF
```

```python
#created a seaborn catplot of the aggreagted data 
jaxDenverCatplot = sb.catplot(data=resetAggGroupedJaxDenverCrimeDF,x='report_year',y='crimes_percapita',aspect=2,kind="point",hue="agency_jurisdiction").set_xticklabels(rotation=90)
```

# Denver, CO
# Unemployment Rate 2012-2021

```python
#Labor Statistics API
#key which I didn't have to use
laborStatisticsKey = "4832da13c89b4a64935299358ebe047b"

#I created a header variable and data variable
#The headers tell the server to make a call, while the data tells the server what to call
headers = {'Content-type': 'application/json'}
data = json.dumps({"seriesid": ['LAUCA082160000000003',],"startyear":"2012", "endyear":"2022"})

#.post allows me to make a request with the data and header varibale
p = requests.post('https://api.bls.gov/publicAPI/v2/timeseries/data/', data=data, headers=headers)

#i use json.loads to conver the json returned by the api to a dictionary
json_data = json.loads(p.text)
```

```python
#I took the data from json_data and created a data frame with pandas 
denverLaborDF = pd.DataFrame(json_data["Results"]["series"][0]['data'])

#I needed to plot this data with seaborn so I converted the data into numeric data instead of strings
denverLaborDF['value'] = pd.to_numeric(denverLaborDF['value'])
denverLaborDF['year'] = pd.to_numeric(denverLaborDF['year'])

#I needed to create column names, remove the footnote section, and add a city column
denverLaborDF.columns = ["Year", "Month", "Month_Name",'Unemployment_Rate_%','footnote']
del denverLaborDF['footnote']
denverLaborDF['City'] = "Denver, CO"
```

```python
denverLaborDF
```

```python
#for this catplot I wanted to use "bar" to show the variance of the unemployment 
unemploymentDenver = sb.catplot(data=denverLaborDF,x='Year',y='Unemployment_Rate_%',
                          aspect=2,kind="bar",hue='City',palette='Blues')
unemploymentDenver
```

```python
#for this seaborn graph I used lineplot with year as x and unemployment as y. Used palette to change the color to blue
unemploymentDenverLineplot = sb.lineplot(data=denverLaborDF,x='Year',y='Unemployment_Rate_%',hue="City",palette='Blues')
unemploymentDenverLineplot
```

# Jacksoville, FL
# Population, Density, and Average Age

```python
#used pandas to read the census csv
popDF = pd.read_csv("CensusPopulation.csv")

#created two variables for Denver and Jacksonville's land area
jaxLandAreaSqMi = 875
denLandAreaSqMi = 154.7
```

```python
popDF
```

```python
# created a data frame with city, year, and population queried by Denver, CO
denPopDF = popDF[["city","year","population"]].query("city=='Denver, CO'")

#created a column in denPopDF for the land area
denPopDF["landArea"] = denLandAreaSqMi

#converted the population to a numeric so I can make calculations
denPopDF["population"] = pd.to_numeric(denPopDF["population"])

#created a new column in the data frame where population/sqMi rounded to the second decimal place
denPopDF["popDensity"] = round(denPopDF["population"]/denLandAreaSqMi,2)
denPopDF
```

```python
# created a data frame with city, year, and population queried by Denver, CO
jaxPopDF = popDF[["city","year","population"]].query("city=='Jacksonville, FL'")

#created a column in denPopDF for the land area
jaxPopDF["landArea"] = jaxLandAreaSqMi

#converted the population to a numeric so I can make calculations
jaxPopDF["population"] = pd.to_numeric(jaxPopDF["population"])

#created a new column in the data frame where population/sqMi rounded to the second decimal place
jaxPopDF["popDensity"] = round(jaxPopDF["population"]/jaxLandAreaSqMi,2)
jaxPopDF
```

```python
#concatenated the tables 
jaxDenPopDensityAgg = pd.concat([jaxPopDF,denPopDF])
jaxDenPopDensityAgg
```

```python
#at the end I figured out how to change to specific colors
palette = {"Jacksonville, FL":"tab:orange","Denver, CO":"tab:blue"}

#created a catplot barchart year on x and population density on y
jaxDenPopCatplot = sb.catplot(data=jaxDenPopDensityAgg,
                                           x="year",y="popDensity",hue="city",kind="bar",aspect=2,palette=palette)
```
